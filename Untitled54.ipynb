```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import spacy

# Assuming your dataframe is named ecommerce_reviews_df
reviews = ecommerce_reviews_df['Review'].dropna().astype(str)

# Load spaCy model for English language
nlp = spacy.load('en_core_web_sm')

# Function to extract named entities (brands) from a text
def extract_entities(text):
    doc = nlp(text)
    entities = [ent.text for ent in doc.ents if ent.label_ == 'ORG']
    return entities

# Extract mentions of competitors from reviews
competitor_mentions = reviews.apply(extract_entities)

# Flatten the list of mentions
competitor_mentions_flat = [mention for mentions_list in competitor_mentions for mention in mentions_list if 'leffe' not in mention.lower()]

# Count the frequency of each mention
mention_counts = Counter(competitor_mentions_flat)

# Plot a bar chart for the most mentioned competitors
plt.figure(figsize=(12, 6))
top_competitors = mention_counts.most_common(10)  # Change the number based on your preference
competitor_names, mention_freq = zip(*top_competitors)
plt.bar(competitor_names, mention_freq, color='skyblue')
plt.title('Top Mentioned Competitors in E-Commerce Reviews')
plt.xlabel('Competitor Brands')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()

# Generate WordCloud for competitor mentions
wordcloud_competitors = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(mention_counts)

# Plotting WordCloud
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud_competitors, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Competitor Mentions in E-Commerce Reviews')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import spacy

# Assuming your dataframe is named ecommerce_reviews_df
reviews = social_listening_df['Social Mention/Comment'].dropna().astype(str)

# Load spaCy model for English language
nlp = spacy.load('en_core_web_sm')

# Function to clean text and remove unwanted characters
def clean_text(text):
    # Remove special characters, URLs, and mentions
    cleaned_text = ' '.join(word for word in text.split() if not (word.startswith('@') or word.startswith('http')))
    return cleaned_text

# Clean the reviews
cleaned_reviews = reviews.apply(clean_text)

# Function to extract named entities (brands) from a text
def extract_entities(text):
    doc = nlp(text)
    entities = [ent.text for ent in doc.ents if ent.label_ == 'ORG']
    return entities

# Extract mentions of competitors from cleaned reviews
competitor_mentions = cleaned_reviews.apply(extract_entities)

# Flatten the list of mentions
competitor_mentions_flat = [mention for mentions_list in competitor_mentions for mention in mentions_list if 'leffe' not in mention.lower()]

# Count the frequency of each mention
mention_counts = Counter(competitor_mentions_flat)

# Plot a bar chart for the most mentioned competitors
plt.figure(figsize=(12, 6))
top_competitors = mention_counts.most_common(10)  # Change the number based on your preference
competitor_names, mention_freq = zip(*top_competitors)
plt.bar(competitor_names, mention_freq, color='skyblue')
plt.title('Top Mentioned Competitors in Social Mention/Comment Reviews')
plt.xlabel('Competitor Brands')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()

# Generate WordCloud for competitor mentions
wordcloud_competitors = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(mention_counts)

# Plotting WordCloud
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud_competitors, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Competitor Mentions in Social Mention/Comment Reviews')
plt.show()
```

```python
import pandas as pd
from textblob import TextBlob  # Assuming you have the TextBlob library installed

# Assuming your e-commerce dataframe is named ecommerce_reviews_df

# Replace 'Your_Text_Column_Name' with the correct column name containing the text of the reviews
ruby_reviews = ecommerce_reviews_df[ecommerce_reviews_df['Flavour Name'] == 'Ruby'][['Review']]
blonde_reviews = ecommerce_reviews_df[ecommerce_reviews_df['Flavour Name'] == 'Blonde'][['Review']]
brune_reviews = ecommerce_reviews_df[ecommerce_reviews_df['Flavour Name'] == 'Brune'][['Review']]

# Function to remove mentions of "Leffe" from text
def remove_leffe_mentions(text):
    return ' '.join(word for word in text.split() if 'leffe' and 'beer' not in word.lower())

# Apply function to remove "Leffe" mentions from each flavor's reviews
ruby_reviews['Review'] = ruby_reviews['Review'].apply(remove_leffe_mentions)
blonde_reviews['Review'] = blonde_reviews['Review'].apply(remove_leffe_mentions)
brune_reviews['Review'] = brune_reviews['Review'].apply(remove_leffe_mentions)

# Function for sentiment analysis using TextBlob
def analyze_sentiment(text):
    analysis = TextBlob(str(text))
    return 'Positive' if analysis.sentiment.polarity > 0 else 'Neutral' if analysis.sentiment.polarity == 0 else 'Negative'

# Apply sentiment analysis to each flavor's reviews
ruby_reviews['Sentiment'] = ruby_reviews['Review'].apply(analyze_sentiment)
blonde_reviews['Sentiment'] = blonde_reviews['Review'].apply(analyze_sentiment)
brune_reviews['Sentiment'] = brune_reviews['Review'].apply(analyze_sentiment)

# Display sentiment distribution for each flavor
print("Ruby Sentiment Distribution:")
print(ruby_reviews['Sentiment'].value_counts())

print("\nBlonde Sentiment Distribution:")
print(blonde_reviews['Sentiment'].value_counts())

print("\nBrune Sentiment Distribution:")
print(brune_reviews['Sentiment'].value_counts())
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Assuming you have the sentiment-tagged dataframes for each flavor (ruby_reviews, blonde_reviews, brune_reviews)
# Replace 'Review' with the actual column name containing the text of the reviews

# Function to remove mentions of "Leffe" and "beer" from text
def remove_mentions(text, excluded_words):
    return ' '.join(word for word in text.split() if all(excluded_word.lower() not in word.lower() for excluded_word in excluded_words))

# Define excluded words
excluded_words = ['Leffe', 'beer','ruby','blonde','brune']

# Apply function to remove excluded words from each flavor's reviews
ruby_reviews['Review'] = ruby_reviews['Review'].apply(lambda x: remove_mentions(x, excluded_words))
blonde_reviews['Review'] = blonde_reviews['Review'].apply(lambda x: remove_mentions(x, excluded_words))
brune_reviews['Review'] = brune_reviews['Review'].apply(lambda x: remove_mentions(x, excluded_words))

# Concatenate reviews for each flavor
all_ruby_reviews = ' '.join(ruby_reviews['Review'].dropna())
all_blonde_reviews = ' '.join(blonde_reviews['Review'].dropna())
all_brune_reviews = ' '.join(brune_reviews['Review'].dropna())

# Generate Word Clouds
wordcloud_ruby = WordCloud(width=800, height=400, background_color='white').generate(all_ruby_reviews)
wordcloud_blonde = WordCloud(width=800, height=400, background_color='white').generate(all_blonde_reviews)
wordcloud_brune = WordCloud(width=800, height=400, background_color='white').generate(all_brune_reviews)

# Plot Word Clouds
plt.figure(figsize=(12, 6))

plt.subplot(131)
plt.imshow(wordcloud_ruby, interpolation='bilinear')
plt.title('Word Cloud for Ruby Reviews')
plt.axis('off')

plt.subplot(132)
plt.imshow(wordcloud_blonde, interpolation='bilinear')
plt.title('Word Cloud for Blonde Reviews')
plt.axis('off')

plt.subplot(133)
plt.imshow(wordcloud_brune, interpolation='bilinear')
plt.title('Word Cloud for Brune Reviews')
plt.axis('off')

plt.show()
```

```python
from wordcloud import STOPWORDS

# Function to generate word clouds and print most repeated words after excluding stop words
def generate_word_cloud_and_print_most_common(reviews, title, num_words=10):
    if reviews:
        # Generate word cloud
        wordcloud = WordCloud(width=800, height=400, random_state=42, max_font_size=80, stopwords=STOPWORDS).generate(reviews)
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title(title)
        plt.show()

        # Print most repeated words after excluding stop words
        stop_words = set(STOPWORDS)
        word_list = [word.lower() for word in reviews.split() if word.lower() not in stop_words]
        word_freq = {word: word_list.count(word) for word in set(word_list)}
        most_common_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:num_words]

        print(f"\nMost Repeated Words (excluding stop words) in {title}:")
        for word, freq in most_common_words:
            print(f"{word}: {freq} times")
    else:
        print(f"No data available for word cloud: {title}")
    

# Assuming you have separate dataframes for positive and negative reviews for each flavor
positive_ruby_reviews = ruby_reviews[ruby_reviews['Sentiment'] == 'Positive']['Review'].str.cat(sep=' ')
negative_ruby_reviews = ruby_reviews[ruby_reviews['Sentiment'] == 'Negative']['Review'].str.cat(sep=' ')

positive_blonde_reviews = blonde_reviews[blonde_reviews['Sentiment'] == 'Positive']['Review'].str.cat(sep=' ')
negative_blonde_reviews = blonde_reviews[blonde_reviews['Sentiment'] == 'Negative']['Review'].str.cat(sep=' ')

positive_brune_reviews = brune_reviews[brune_reviews['Sentiment'] == 'Positive']['Review'].str.cat(sep=' ')
negative_brune_reviews = brune_reviews[brune_reviews['Sentiment'] == 'Negative']['Review'].str.cat(sep=' ')

# Generate word clouds and print most repeated words (excluding stop words) for positive reviews
generate_word_cloud_and_print_most_common(positive_ruby_reviews, 'Positive Reviews for Ruby Flavor')
generate_word_cloud_and_print_most_common(positive_blonde_reviews, 'Positive Reviews for Blonde Flavor')
generate_word_cloud_and_print_most_common(positive_brune_reviews, 'Positive Reviews for Brune Flavor')

# Generate word clouds and print most repeated words (excluding stop words) for negative reviews
generate_word_cloud_and_print_most_common(negative_ruby_reviews, 'Negative Reviews for Ruby Flavor')
generate_word_cloud_and_print_most_common(negative_blonde_reviews, 'Negative Reviews for Blonde Flavor')
generate_word_cloud_and_print_most_common(negative_brune_reviews, 'Negative Reviews for Brune Flavor')
```
