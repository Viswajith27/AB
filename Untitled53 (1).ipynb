```python
# Group 1: Queries and Interest_Queries
group1_cols = ['Queries', 'Interest_Queries']
group1_df = google_search_df[group1_cols]

# Group 2: Week and Interest_Week
group2_cols = ['Week', 'Interest_Week']
group2_df = google_search_df[group2_cols]

# Group 3: Region and Interest_Region
group3_cols = ['Region', 'Interest_Region']
group3_df = google_search_df[group3_cols]

# Group 4: Keyword and AvgMonthlySearches
group4_cols = ['Keyword', 'AvgMonthlySearches']
group4_df = google_search_df[group4_cols]

# Display the first few rows of each group to verify
print("Group 1:")
print(group1_df.head())
print("\nGroup 2:")
print(group2_df.head())
print("\nGroup 3:")
print(group3_df.head())
print("\nGroup 4:")
print(group4_df.head())
```

```python
# Convert "<1" to NaN in the Interest_Queries column
group1_df['Interest_Queries'] = pd.to_numeric(group1_df['Interest_Queries'], errors='coerce')

# Display unique queries and their corresponding interest scores
print("Unique Queries and Interest Scores:")
print(group1_df[['Queries', 'Interest_Queries']].drop_duplicates())

# Visualize the distribution of interest scores
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(data=group1_df, x='Interest_Queries', bins=20, kde=True)
plt.title('Distribution of Interest Scores for Queries')
plt.xlabel('Interest Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Display unique weeks and their corresponding interest scores
print("Unique Weeks and Interest Scores:")
print(group2_df[['Week', 'Interest_Week']].drop_duplicates())

# Visualize the trend of interest scores over weeks with rotated x-axis labels
plt.figure(figsize=(12, 6))
sns.lineplot(x='Week', y='Interest_Week', data=group2_df)
plt.title('Interest Scores Over Weeks')
plt.xlabel('Week')
plt.ylabel('Interest Score')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better visualization
plt.show()
```

```python
# Display unique regions and their corresponding interest scores
print("Unique Regions and Interest Scores:")
print(group3_df[['Region', 'Interest_Region']].drop_duplicates())

# Visualize interest scores by region
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Interest_Region', data=group3_df)
plt.title('Interest Scores by Region')
plt.xlabel('Region')
plt.ylabel('Interest Score')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=group4_df, x='AvgMonthlySearches', order=group4_df['AvgMonthlySearches'].value_counts().index)
plt.title('Distribution of Average Monthly Searches for Keywords')
plt.xlabel('Avg Monthly Searches')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
